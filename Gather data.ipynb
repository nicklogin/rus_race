```python
import os
import pandas as pd
```

```python
train_data = []

for file in os.listdir("data"):
    if "_train_" in file:
        train_data.append(pd.read_csv(f"data/{file}", index_col="Unnamed: 0"))
```

```python
train_data_concat = pd.concat(train_data)
```

```python
train_data_concat
```

```python
os.mkdir("clean_data")
train_data_concat.to_csv("clean_data/race_ru_train.csv")
```
